```python
import matplotlib.pyplot as plt
```

## 1.line chart

```python
x = [2,4,5,6]
y = [10,30,60,120]
```

```python
plt.plot(x,y)
plt.title("line data")
plt.xlabel("x-axis")
plt.ylabel("Y-axis")
plt.savefig("linechart.jpg", dpi = 400)
plt.show()
```

```python
name = ['A', "B", "C", "D"]
age = [20, 30, 17, 61]
```

```python
plt.bar(name, age)
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Bar Chart")
plt.savefig("BarChart.jpg", dpi = 400)
```

```python
name = ['A', "B", "C", "D"]
age = [20, 30, 17, 61]

plt.pie(age, labels=name, autopct="%1.1f%%")
plt.savefig("pieChart.jpg", dpi = 400)
plt.show()
```
